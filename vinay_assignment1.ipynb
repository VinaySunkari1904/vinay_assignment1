```python
import pandas as pd
```

```python
data = pd.read_csv('customer_transactions.csv')
print(data.head())
```

```python
data.shape
```

```python
#checking for missing values
missing_data_columns =  data.isnull().sum()
print(missing_data_columns)
```

```python
data['Quantity_Purchased']=data['Quantity_Purchased'].fillna(data['Quantity_Purchased'].median())
data['Transaction_Value'] =data['Transaction_Value'].fillna(data['Transaction_Value'].median())
```

```python
data.head(27)
```

```python
# Drop rows with more than 50% missing data
data.dropna(thresh=len(data.columns) * 0.5)
data
```

```python
# Drop columns with more than 70% missing data
data = data.loc[:, data.isnull().mean() < 0.7]
data
```

```python
# Outlier detection and handling
def replace_outliers_with_median(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_value = df[column_name].median()
    df.loc[(df[column_name] < lower_bound) | (df[column_name] > upper_bound), column_name] = median_value
```

```python
#outliers for 'Transaction_Value'
replace_outliers_with_median(data, 'Transaction_Value')
```

```python
#extreme outliers and negative values for 'Delivery_Time'
delivery = data[data['Delivery_Time'] >= 0]
replace_outliers_with_median(delivery, 'Delivery_Time')
```

```python
### Categorical Variable Encoding
# Convert Payment_Method and Is_Returned into categorical variables.
payment_dummies = pd.get_dummies(data['Payment_Method'], prefix='Payment')
data = pd.concat([data, payment_dummies], axis=1)
```

```python
# Create a binary variable Return_Flag for Is_Returned 
data['Is_Returned'] = data['Is_Returned'].astype('category')
data['Return_Flag'] = data['Is_Returned'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
#new feature Transaction_Year by extracting the year from Transaction_Date.
data['Transaction_Date'] = pd.to_datetime(data['Transaction_Date'])
data['Transaction_Year'] = data['Transaction_Date'].dt.year
```

```python
# Create 'High_Rating' feature
data['High_Rating'] = data['Customer_Rating'].apply(lambda x: 1 if x >= 4 else 0)
```

```python
data.to_csv('cleaned_customer_transactions.csv', index=False)
```
